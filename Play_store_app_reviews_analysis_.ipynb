# **Exploration of data**



   
#### 1. How many apps registered in each category ?




```python
# Grouping of data by Category and then count the values of each category
registered_apps_by_category=pd.DataFrame(play_store_data['Category'].value_counts())
```

```python
registered_apps_by_category.columns=['No of Apps']
```

```python
registered_apps_by_category
```

```python
# bar chart for registered apps by category 
plt.figure(figsize=(14,9))
plt.barh(registered_apps_by_category.index,registered_apps_by_category['No of Apps'])
plt.title(" Registered apps by category")
plt.xlabel("No of apps")
plt.ylabel("Category")
plt.show
```

#### 2. Top ten most reviewed apps in Family category ?





```python
# Grouping the dataframe by Category accessing the group named family sorting value by reviews in descending order limit it by 10 from the top
play_store_data.groupby('Category').get_group('FAMILY').sort_values(by=['Reviews'],ascending=False).head(10)
```

#### 3. Which app is most installed in each category ?

```python
# Grouping the data by Category access the installs column finding max index value then access three column named App,Category,Installs
play_store_data.loc[play_store_data.groupby('Category')['Installs'].idxmax(),['App','Category','Installs']]
```

#### 4. On the whole what is the percentage of content rating of apps ?


```python
#Group the data frame by content rating access the column content rating count the values
play_store_content_rating=pd.DataFrame(play_store_data.groupby('Content Rating')['Content Rating'].count())
```

```python
# Rename the column by Value count
play_store_content_rating.columns=['Value count']
```

```python
# finding the percentage of value count by rounding of to 2 digit
play_store_content_rating['percentage']=round(play_store_content_rating['Value count']*100/play_store_content_rating['Value count'].sum(),2)
```

```python
play_store_content_rating
```

```python
# plotting the pie chart for play store content rating percentage
plt.figure(figsize=(10,14))
plt.pie(play_store_content_rating['percentage'])
plt.legend(play_store_content_rating.index)

plt.title("play store content rating percentage")
```

#### 5. How many apps are updated in each year ?

```python
play_store_data.head(2)
```

```python
# Accessing the year from column name last updated
play_store_data['update year']=pd.DatetimeIndex(play_store_data['Last Updated']).year
```

```python
# Grouping data by year and count values in each year
play_store_app_udated_by_year=pd.DataFrame(play_store_data.groupby('update year')['update year'].count())
```

```python
# Rename the column 
play_store_app_udated_by_year.columns=['No of apps']
```

```python
print(play_store_app_udated_by_year)
```

```python
# plot Year on year app updated 
plt.plot(play_store_app_udated_by_year.index,play_store_app_udated_by_year['No of apps'])
plt.title("Year on year app updated")
plt.xlabel('year')
plt.ylabel("no of apps")
plt.show()
```
